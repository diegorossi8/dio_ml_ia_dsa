```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Carrega a imagem da Lena do diretrio local
img = cv2.imread('lenna.jpg')

# Verifica se a imagem foi carregada corretamente
if img is None:
    print("No foi possvel carregar a imagem.")
else:
    print("Imagem carregada com sucesso!")

    # Converte a imagem para nveis de cinza
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Converte a imagem para binarizada (preto e branco)
    _, binary_img = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)

    # Exibe as 3 imagens
    cv2_imshow(img)
    cv2_imshow(gray_img)
    cv2_imshow(binary_img)

    # Aguarda o usurio pressionar uma tecla para fechar as janelas
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
